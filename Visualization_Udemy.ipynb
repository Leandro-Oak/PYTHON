##Intro to Seaborn - It is built on top of Matplotlib

##Stacked Distribution / Histograms

```python
#version 1
#Graph with filters of two columns
#And show them on top of each other


plt.hist(movies[movies['Genre']=="Action"].BudgetMillions,bins=15 )
plt.hist(movies[movies['Genre']=="Drama"].BudgetMillions,bins=15)
plt.hist(movies[movies['Genre']=="Romance"].BudgetMillions,bins=15)
plt.show()
```

```python
#Version 2
#Side side of each other

plt.hist([movies[movies['Genre']=="Action"].BudgetMillions,\
         movies[movies['Genre']=="Drama"].BudgetMillions,\
         movies[movies['Genre']=="Romance"].BudgetMillions],bins=15)
plt.show()
```

```python
#Version 3
#Stacked on top of each other


plt.hist([movies[movies['Genre']=="Action"].BudgetMillions,\
         movies[movies['Genre']=="Drama"].BudgetMillions],bins=15, stacked = True,)
plt.show()
```

```python
#Version 4
# Using For looping to define multiple Genre
filter= []
mylabels = []
for x in movies['Genre'].cat.categories:
  filter.append(movies[movies['Genre']== x].BudgetMillions)
  mylabels.append(x)


h = plt.hist(filter, bins=30, stacked=True, rwidth=3, label=mylabels)
plt.legend()
plt.show()
```
